```python
import numpy as np
import pandas as pd
import scipy

from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram,linkage
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import pickle
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_segmentation=pd.read_csv('segmentation data.csv',index_col=0)#index_col=0 means that the 1st column of the dataset is treated as index column
```

```python
df_segmentation.head(5)
```

```python
df_segmentation.describe()
```

## Correlation Estimate

```python
df_segmentation.corr()
```

```python
plt.figure(figsize=(12,9))
s=sns.heatmap(df_segmentation.corr(),
             annot=True,
             cmap='RdBu',
             vmin=-1, 
             vmax=1)
s.set_yticklabels(s.get_yticklabels(),rotation=0,fontsize=12)
s.set_xticklabels(s.get_xticklabels(),rotation=90,fontsize=12)
plt.title('Correlation Heatmap')
plt.show()
```

## Visualize Raw Data

```python
plt.figure(figsize=(12,9))
plt.scatter(df_segmentation.iloc[:,2],df_segmentation.iloc[:,4])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Visualization of Raw Data')
```

## Standardization

```python
scaler=StandardScaler()
segmentation_std=scaler.fit_transform(df_segmentation)
```

## Hierachical Clustering

```python
hier_clust=linkage(segmentation_std,method='ward')
```

```python
plt.figure(figsize=(12,9))
plt.title('Hierachical Clustering Dendogram')
plt.xlabel('Observations')
plt.ylabel('Distance')
dendrogram(hier_clust,
         truncate_mode='level',
         p=5,
         show_leaf_counts=False)
         #color_threshold=0)
plt.show()
```

## K-Means Clustering

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)#k-means is the algorithm that finds the best centroids
    kmeans.fit(segmentation_std)
    wcss.append(kmeans.inertia_)#within cluster sqd. sum is stored in inertia attribute
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,11),wcss,marker='o',linestyle='--')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('K-means Clustering')
plt.show()
```

```python
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)#k-means is the algorithm that finds the best centroids
kmeans.fit(segmentation_std)
```

## Results

```python
df_segm_kmeans=df_segmentation.copy()
df_segm_kmeans['Segment k-means']=kmeans.labels_
```

```python
df_segm_analysis=df_segm_kmeans.groupby(['Segment k-means']).mean()
df_segm_analysis
```

```python
df_segm_analysis['N Obs']=df_segm_kmeans[['Segment k-means','Sex']].groupby(['Segment k-means']).count()#we need to have a second argument in order to count the number of rows it can be anything here 'sex' is taken
```

```python
df_segm_analysis.head(5)
```

```python
df_segm_analysis['Prop Obs']=df_segm_analysis['N Obs']/df_segm_analysis['N Obs'].sum()
```

```python
df_segm_analysis.rename({0:'well-off',
                         1:'fewer opportunities',
                         2:'standard',
                         3:'career focused'})
```

```python
df_segm_kmeans['Labels']=df_segm_kmeans['Segment k-means'].map({0:'well-off',
                         1:'fewer opportunities',
                         2:'standard',
                         3:'career focused'})
```

```python

x_axis=df_segm_kmeans['Age']
y_axis=df_segm_kmeans['Income']
plt.figure(figsize=(10,8))
sns.scatterplot(x_axis,y_axis,hue=df_segm_kmeans['Labels'],palette=['g','r','c','m'])
plt.title('Segmentation K-means')
plt.show()
```

## PCA

```python
pca=PCA()
```

```python
pca.fit(segmentation_std)
```

```python
pca.explained_variance_ratio_ #as there are 7 features so 7 components
```

```python
plt.figure(figsize=(12,9))
plt.plot(range(1,8),pca.explained_variance_ratio_.cumsum(),marker='o',linestyle='--')
plt.title('Explained variance by Components')
plt.xlabel('Number of Components')
plt.ylabel('CUmumlative explained variance')
```

```python
pca=PCA(n_components=3)
```

```python
pca.fit(segmentation_std)
```

## PCA results

```python
pca.components_
```

```python
df_pca_comp=pd.DataFrame(data=pca.components_,columns=df_segmentation.columns.values,index=['Compoenet 1','Compoenet 2','Compoenet 3'])
df_pca_comp
```

```python
plt.figure(figsize=(12,9))
sns.heatmap(df_pca_comp,
             annot=True,
             cmap='RdBu',
             vmin=-1, 
             vmax=1)
plt.yticks([0,1,2],['Compoenet 1','Compoenet 2','Compoenet 3'],rotation=45,fontsize=9)
#this graph is very important we see which components has most correlated features and those will be dominant features in those components
#componet 1 has income occupation settlement size as its highest correlated features so component 1 denotes career
#componet 2 has education martital status and sex as its highest correlated features so component 2 denotes educational lifestyle
#componet 3 has martital status age occupation as its highest correlated features so component 3 denotes educational experience
```

```python
pca.transform(segmentation_std)#we convert for 7 dimensional to 3 dimensional
```

```python
scores_pca=pca.transform(segmentation_std)
```

## K-Means clustering with pca

```python
wcss=[]
for i in range(1,11):
    kmeans_pca=KMeans(n_clusters=i,init='k-means++',random_state=42)#k-means is the algorithm that finds the best centroids
    kmeans_pca.fit(segmentation_std)
    wcss.append(kmeans_pca.inertia_)#within cluster sqd. sum is stored in inertia attribute
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,11),wcss,marker='o',linestyle='--')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('K-means Clustering')
plt.show()
```

```python
kmeans_pca=KMeans(n_clusters=4,init='k-means++',random_state=42)#k-means is the algorithm that finds the best centroids
```

```python
kmeans_pca.fit(scores_pca)
```

## K-means clustering with PCA results

```python
df_segm_pca_kmeans=pd.concat([df_segmentation.reset_index(drop=True),pd.DataFrame(scores_pca)],axis=1)
df_segm_pca_kmeans.columns.values[-3:]=['Component 1','Component 2','Component 3']
df_segm_pca_kmeans['Segemnt k-means PCA']=kmeans_pca.labels_
```

```python
df_segm_pca_kmeans
```

```python
df_segm_pca_kmeans_freq=df_segm_pca_kmeans.groupby(['Segemnt k-means PCA']).mean()
df_segm_pca_kmeans_freq
#component 1 denote career comp 2 educational lifestyle comp 3 experience
#Segemnt k-means PCA value 3 has high in all comp so it is well off
#Segemnt k-means PCA value 2 has lowest average in comp 2 and 1, and 3 is high so it is fewer opportunities cluster
#Segemnt k-means PCA value 1 has high in career comp i.e 1 so it is career focused cluster
#Segemnt k-means PCA value 0 has low in 2 and OK in 1 and 3 so it is standard cluster
```

```python
df_segm_pca_kmeans_freq['N Obs']=df_segm_pca_kmeans[['Segemnt k-means PCA','Sex']].groupby(['Segemnt k-means PCA']).count()#we need to have a second argument in order to count the number of rows it can be anything here 'sex' is taken
df_segm_pca_kmeans_freq['Prop Obs']=df_segm_pca_kmeans_freq['N Obs']/df_segm_pca_kmeans_freq['N Obs'].sum()
df_segm_pca_kmeans_freq=df_segm_pca_kmeans_freq.rename({3:'well-off',
                         2:'fewer opportunities',
                         0:'standard',
                         1:'career focused'})
df_segm_pca_kmeans_freq
```

```python
df_segm_pca_kmeans['Legend']=df_segm_pca_kmeans['Segemnt k-means PCA'].map({3:'well-off',
                         2:'fewer opportunities',
                         0:'standard',
                         1:'career focused'})
```

```python
x_axis=df_segm_pca_kmeans['Component 2']
y_axis=df_segm_pca_kmeans['Component 1']
plt.figure(figsize=(10,8))
sns.scatterplot(x_axis,y_axis,hue=df_segm_pca_kmeans['Legend'],palette=['g','r','c','m'])
plt.title('Culsters by PCA componenets')
plt.show()
```

```python
x_axis=df_segm_pca_kmeans['Component 3']
y_axis=df_segm_pca_kmeans['Component 2']
plt.figure(figsize=(10,8))
sns.scatterplot(x_axis,y_axis,hue=df_segm_pca_kmeans['Legend'],palette=['g','r','c','m'])
plt.title('Culsters by PCA componenets')
plt.show()
```

```python
x_axis=df_segm_pca_kmeans['Component 3']
y_axis=df_segm_pca_kmeans['Component 1']
plt.figure(figsize=(10,8))
sns.scatterplot(x_axis,y_axis,hue=df_segm_pca_kmeans['Legend'],palette=['g','r','c','m'])
plt.title('Culsters by PCA componenets')
plt.show()
```
